### **Exploratory Data Analysis of E-Commerce Sales**

### **1.  Analyze the monthly sales of the store and identifying which month had the highest and lowest sales.**

```python
sales_by_month = data.groupby('order month')['Sales'].sum().reset_index()
sales_by_month

                              
```

```python
fig = px.line(sales_by_month, 
                x='order month', 
                y='Sales', 
                title='Monthly Sales Analysis',
                labels={'order month': 'Month', 'Sales': 'Total Sales'},
                markers=True)
fig.show()
```

### **2. Analyze Sales based on product category and determine which category had highest sales or lowest sales**

```python
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
sales_by_category 
```

```python
fig = px.pie(sales_by_category,
             values='Sales',
             names='Category',
             hole=0.5,
             color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Sales Analysis by Category', title_font = dict(size = 24))

fig.show()
```

### **3.The Sales Analysis based on  Sub-Category.**

```python
sales_by_subcategory = data.groupby('Sub-Category')['Sales'].sum().reset_index()
sales_by_subcategory
```

```python
fig = px.bar(sales_by_subcategory, x='Sub-Category', y='Sales', title='Sales Analysis by Sub-Category')
fig.show()
```

### **4.What is the monthly profit from sales and determine which month had the highest profit?**

```python
# Convert to datetime
data['Order Date'] = pd.to_datetime(data['Order Date'], errors='coerce')

# Extract month-year and store as string (avoids JSON serialization issues)
data['Order Month'] = data['Order Date'].dt.to_period('M').astype(str)

# Group and sum profit
profit_by_month = (
    data.groupby('Order Month')['Profit']
    .sum()
    .reset_index()
    .sort_values('Order Month')
)

profit_by_month

```

```python
fig = px.bar(profit_by_month, x='Order Month', y='Profit', title='Monthly Profit Analysis')
fig.show()
```

### **5. Profit By Category**

```python
profit_by_category = data.groupby('Category')['Profit'].sum().reset_index()
profit_by_category
```

```python
fig = px.pie(profit_by_category,
             values='Profit',
             names='Category',
             hole=0.5,
             color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Profit Analysis by Category', title_font = dict(size = 24))

fig.show()
```

### **6.  Profit by sub-category**

```python
profit_by_subcategory = data.groupby('Sub-Category')['Profit'].sum().reset_index()
fig = px.bar(profit_by_subcategory, 
             x='Sub-Category', 
             y='Profit', 
             title='Profit Analysis by Sub-Category')
fig.show()
```

### **7. Analyze the Sales Profit - By Segment**

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

color_palette = px.colors.qualitative.Pastel

fig = go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Sales'],
                     name='Sales',
                     marker_color=color_palette[0]))

fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Profit'],
                     name='Profit',
                     marker_color=color_palette[1]))

fig.update_layout(title='Sales and Profit Analysis by customer Segment',
                  xaxis_title='customer Segment', yaxis_title='Amount')

fig.show()
```

### **8. Analye the Sales to Profit Ratio.**

```python


# Avoid division by zero by replacing 0 profit with NaN
data['sales_to_profit_ratio'] = np.where(
    data['Profit'] != 0,
    data['Sales'] / data['Profit'],
    np.nan
)

data[['Sales', 'Profit', 'sales_to_profit_ratio']].head()
```

### **9.  Group by Segment and sum both Sales and Profit**

```python
# Group by Segment and sum both Sales and Profit
sales_profit_by_segment = (
    data.groupby('Segment')[['Sales', 'Profit']]
    .sum()
    .reset_index()
)

# Calculate ratio
sales_profit_by_segment['Sales_to_Profit_Ratio'] = (
    sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
)

# Show only Segment and the ratio
print(sales_profit_by_segment[['Segment', 'Sales_to_Profit_Ratio']])
```

### **10. what is the Top 10 most profitable products ?**

```python
# Group by Product Name and sum the Profit
top_products = data.groupby('Product Name')['Profit'].sum()

# Sort in descending order and select the top 10
top_products = top_products.sort_values(ascending=False).head(10).reset_index()

print(top_products)


fig = px.bar(top_products, x='Product Name', y='Profit',
             title='Top 10 Most Profitable Products',
             text='Profit')
fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig.show()
```

### **11. What is the Top 10 loss-making products (lowest total profit) ?**

```python
# Group by Product Name and sum the Profit
loss_products = data.groupby('Product Name')['Profit'].sum()

# Filter only products with negative profit
loss_products = loss_products[loss_products < 0]

# Sort in ascending order (most loss at top) and select top 10
loss_products = loss_products.sort_values().head(10).reset_index()

print(loss_products)

fig = px.bar(loss_products, x='Product Name', y='Profit',
             title='Top 10 Loss-Making Products',
             text='Profit')
fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig.show()
```

### **12. How much profit gained by shipping mode ?**

```python
# Group by Ship Mode and sum the Profit
profit_by_ship = data.groupby('Ship Mode')['Profit'].sum().reset_index()

print(profit_by_ship)

# Optional: Plot with Plotly

fig = px.bar(profit_by_ship, x='Ship Mode', y='Profit',
             title='Profit by Shipping Mode',
             text='Profit',
             color='Profit')
fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig.show()
```

### **13. Who is our Customer Lifetime Value (CLV) ?**

```python
# Group by Customer Name and sum the Sales

clv = data.groupby('Customer Name')['Sales'].sum().reset_index()
clv = clv.sort_values(by='Sales', ascending=False)
fig1 = px.bar(clv.head(10), x='Customer Name', y='Sales',
              title="Top 10 Customers by Lifetime Value",
              text='Sales')
fig1.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig1.show()
```

### **14. What is the monthly Sales Trend ?**

```python
monthly_sales = data.groupby(pd.Grouper(key='Order Date', freq='ME'))['Sales'].sum().reset_index()
fig2 = px.line(monthly_sales, x='Order Date', y='Sales',
               title="Monthly Sales Trend",
               markers=True)
fig2.show()
```

### **15. What is the average day to ship by shipping mode ?**

```python

# Ensure dates are in datetime format
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])

# Create Days to Ship column
data['Days to Ship'] = (data['Ship Date'] - data['Order Date']).dt.days
delivery_perf = data.groupby('Ship Mode')['Days to Ship'].mean().reset_index()
fig3 = px.bar(delivery_perf, x='Ship Mode', y='Days to Ship',
              title="Average Days to Ship by Shipping Mode",
              text='Days to Ship', color='Days to Ship')
fig3.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig3.show()
```

### **16. What's the sales Heatmap by Category and Region ?**

```python
heatmap_data = data.groupby(['Category', 'Region'])['Sales'].sum().reset_index()
fig4 = px.density_heatmap(heatmap_data, x='Category', y='Region', z='Sales',
                          title="Sales Heatmap by Category and Region",
                          text_auto=True, color_continuous_scale='Blues')
fig4.show()
```

### **20.  Pareto Analysis (80/20 Rule)**

```python
pareto = data.groupby('Product Name')['Sales'].sum().reset_index()
pareto = pareto.sort_values(by='Sales', ascending=False).reset_index(drop=True)
pareto['Cumulative Sales %'] = pareto['Sales'].cumsum() / pareto['Sales'].sum() * 100

fig5 = px.line(pareto, y='Cumulative Sales %',
               title="Pareto Analysis (80/20 Rule)",
               markers=True)
fig5.add_hline(y=80, line_dash="dash", line_color="red")
fig5.show()
```

### **21. What is the Order Size Distribution (Quantity per Order) ?**

```python
if 'Order ID' in data.columns and 'Quantity' in data.columns:
    order_qty = data.groupby('Order ID', dropna=False)['Quantity'].sum().reset_index()
    fig = px.histogram(order_qty, x='Quantity', nbins=30, title='Order Size Distribution (Quantity per Order)')
    fig.show()
else:
    print("Skipping Order Size distribution — need Order ID and Quantity.")
```

### Year-over-Year Growth (Sales & Profit)

###  Missing Data

```python
if 'Year' in data.columns and 'Sales' in data.columns and 'Profit' in data.columns:
    yoy = data.groupby('Year').agg({'Sales':'sum','Profit':'sum'}).reset_index()
    yoy['Sales YoY %'] = yoy['Sales'].pct_change() * 100
    yoy['Profit YoY %'] = yoy['Profit'].pct_change() * 100
    fig = px.line(yoy, x='Year', y=['Sales','Profit'], title='Sales & Profit by Year', markers=True)
    fig.show()
    print(yoy)
else:
    print("Skipping YoY growth — required year/sales/profit missing.")
```

### **22. What is the Correlation between Discount, Sales and Profit ?**

```python
corr_cols = [c for c in ['Discount','Sales','Profit'] if c in data.columns]
if len(corr_cols) >= 2:
    corr_df = data[corr_cols].dropna()
    corr_matrix = corr_df.corr()
    print("Correlation matrix:\n", corr_matrix)
    # Heatmap via plotly
    fig = px.imshow(corr_matrix, text_auto=True, title='Correlation Matrix')
    fig.show()
else:
    print("Skipping Correlation analysis — at least two of Discount/Sales/Profit required.")
```
